```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

df = pd.read_csv(r'D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Market Basket Analysis (MBA) & Bundling Strategy\data\clean\clean_dataset.csv')
df
```

```python
# ubah long menjadi list-of-list berdasarkan order id
basket = df.groupby('order_id')['product_name'].apply(list).tolist()

# Inisialisasi TransactionEncoder
te = TransactionEncoder()

# fit dan Transform ke format matriks (gunakan sparse=True agar angka 0 tidak memakan memori RAM)
te_ary = te.fit(basket).transform(basket, sparse=True)

# masukkan ke DataFrame
df_encoded = pd.DataFrame.sparse.from_spmatrix(te_ary, columns=te.columns_)

# Cek hasil
print(f"Jumlah Transaksi: {df_encoded.shape[0]}")
print(f"Jumlah Produk Unik: {df_encoded.shape[1]}")
df_encoded.head()
```
