```python
import pandas as pd
```

#### Join (left) data

```python
aisle = pd.read_csv(r'D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Market Basket Analysis (MBA) & Bundling Strategy\data\raw\aisles.csv')
department = pd.read_csv(r'D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Market Basket Analysis (MBA) & Bundling Strategy\data\raw\departments.csv')
order_product = pd.read_csv(r'D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Market Basket Analysis (MBA) & Bundling Strategy\data\raw\order_products__prior.csv')
order = pd.read_csv(r'D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Market Basket Analysis (MBA) & Bundling Strategy\data\raw\orders.csv')
product = pd.read_csv(r'D:\Dhika Ramadhan Saputra\Data Analist\Full Project\Market Basket Analysis (MBA) & Bundling Strategy\data\raw\products.csv')
```

```python
df_merged = pd.merge(order_product, product, on='product_id', how='left')
df_merged = pd.merge(df_merged, aisle, on='aisle_id', how='left')
df_merged = pd.merge(df_merged, department, on='department_id', how='left')
df_final = pd.merge(df_merged, order, on='order_id', how='left')

print(df_final.head())

print(df_final.info())
```

#### Features engineering

```python
kolom_pilihan = [
    'order_id',
    'user_id',
    'product_id', 
    'product_name', 
    'aisle', 
    'department', 
    'add_to_cart_order', 
    'reordered', 
    'order_dow', 
    'order_hour_of_day', 
    'days_since_prior_order'
]
df_final = df_final[kolom_pilihan]

df_final.head()
```

```python
unique_orders = df_final['order_id'].unique()
orders_sample = unique_orders[:100000] 

df_sample = df_final[df_final['order_id'].isin(orders_sample)]

print(f"Total baris sekarang: {len(df_sample):,}")
print(f"Total pesanan (order_id) unik: {df_sample['order_id'].nunique():,}")
```
