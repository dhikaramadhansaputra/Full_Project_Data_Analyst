### Distribusi data

```python
# Kategorial
# 1 baris dan 3 kolom
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

ax = axes.flatten()

# Plot 1: Category
sns.countplot(data=df, x='Category', ax=ax[0], color='blue')
ax[0].set_title('Category')
ax[0].tick_params(axis='x', rotation=45)

# Plot 2: Payment_Method
sns.countplot(data=df, x='Payment_Method', ax=ax[1], color='blue')
ax[1].set_title('Payment Method')
ax[1].tick_params(axis='x', rotation=45)

# Plot 3: Province
sns.countplot(data=df, x='Province', ax=ax[2], color='blue')
ax[2].set_title('Province')
ax[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the figure and the 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Subplot 2: Just Unit Cost
axes[0].boxplot(df['Unit_Cost'].dropna())
axes[0].set_title('Unit Cost Only')
axes[0].set_xticklabels(['Unit Cost'])

# Subplot 3: Just Revenue
axes[1].boxplot(df['Revenue'].dropna())
axes[1].set_title('Revenue Only')
axes[1].set_xticklabels(['Revenue'])

plt.tight_layout()
plt.show()

# LOGIKANYA SALAH, TANYA AI GIMANA CARA CEK OUTLAYER HARGA
```
